```python
def extract_key_value_pairs(cleaned_list, search_words):
    key_value_pairs = {}
 
    search_words_lower = [word.lower() for word in search_words]

    for item in cleaned_list:
        item = item.strip()
        if not item or item == '-':
            continue
 
        item_lower = item.lower()
 
        for word, word_lower in zip(search_words, search_words_lower):
            if word_lower in item_lower: 
                key_value_pairs[word] = item
    
    return key_value_pairs

# Dictionary   
key_value_pairs = extract_key_value_pairs(extracted_text, search_words)
 
for key, value in key_value_pairs.items():
    print(f"{key}: {value}")
```

```python
def extract_info(text,text2):
    property_information = {}
    
    #defining the regular expression for each column
    property_id_pattern = r'Kiinteistötunnus:\s+\d{3}-\d{3}-\d-\d{3}-\w\d'
    property_information['Property ID']= re.search(property_id_pattern, extracted_text2).group().split()[1] if re.search(property_id_pattern, extracted_text2) else None
    
    #Address_pattern
    Address_pattern1 =  re.compile(r'Osoite\s+([^,]+)')    #address
    Address_pattern2 = re.compile(r'osoitteessa\s+([^,]+)')
    property_information['Address'] = re.search(input2, extracted_text2).group(1) if (input2 := Address_pattern1 if Address_pattern1.search(extracted_text2) else Address_pattern2 if Address_pattern2.search(extracted_text2) else None) else None

    #Municipality and Postal_code pattern
    code_pattern = r'(\b\w{5}\b) (\w+)'   #postal_code
    muncipality_pattern1 = re.compile(r'Osoite\s+([^,]+),([^,]+)') #Address
    muncipality_pattern2 = re.compile(r'osoitteessa\s+([^,]+),([^,]+)')
    muncipality_code = re.search(input3, extracted_text2).group() if (input3 := muncipality_pattern1 if muncipality_pattern1.search(extracted_text2) else muncipality_pattern2 if muncipality_pattern2.search(extracted_text2) else None) else None
    property_information['Municipality'] = re.search(code_pattern, muncipality_code).group(2) if (input3 := muncipality_pattern1 if muncipality_pattern1.search(extracted_text2) else muncipality_pattern2 if muncipality_pattern2.search(extracted_text2) else None) else None
    property_information['Postal code'] = re.search(code_pattern, muncipality_code).group(1) if (input3 := muncipality_pattern1 if muncipality_pattern1.search(extracted_text2) else muncipality_pattern2 if muncipality_pattern2.search(extracted_text2) else None) else None
    
    #plot ownership patterns
    plot_ownership_pattern1 = r'vuokratontilla' #ownership pattern
    plot_ownership_pattern2 = r'omalla tontilla'
    plot_ownership_pattern3 = r'oma tontti'
    property_information['Plot ownership'] = 'Freehold' if re.search(plot_ownership_pattern2, extracted_text2) or re.search(plot_ownership_pattern3, extracted_text2) else 'Leasehold'

    #property_type
    #type_of_property_pattern = r'Omakotitalo' #type of property pattern
    type_of_property_pattern1 = r'Kiinteistön'
    type_of_property_pattern2 = r'asuinhuoneisto'
    type_of_property_pattern3 = r'rivitalo'
    type_of_property_pattern4 = r'erillistalo'
    type_of_property_pattern5 = r'paritalo'

    if re.search(type_of_property_pattern1, extracted_text2):
        property_information['Type of property'] = 'Single family house'
    elif re.search(type_of_property_pattern2, extracted_text2):
        property_information['Type of property'] = 'Apartment'
    elif re.search(type_of_property_pattern3, extracted_text2):
        property_information['Type of property'] = 'Rowhouse'
    elif re.search(type_of_property_pattern4, extracted_text2):
        property_information['Type of property'] = 'detached house'
    elif re.search(type_of_property_pattern5, extracted_text2):
        property_information['Type of property'] = 'dublex'
    else:
        property_information['Type of property'] = None
    
    property_information['Type Code'] = None
    value_year_pattern = r'arvostettu\s+arvostettu(\d{4})'  #value year & building year pattern
    property_information['Value year'] = re.search(value_year_pattern,extracted_text2).group(1) if re.search(value_year_pattern,extracted_text2) else None
    #living area pattern
    living_area_pattern1 = r'Huoneistoala (\d+)m2'
    living_area_pattern2 = r'pinta-ala on (\d+)m2'

    if re.search(living_area_pattern1,extracted_text2):
        property_information['Living area'] = re.search(living_area_pattern1,extracted_text2).group(1)
    elif re.search(living_area_pattern2,extracted_text2):
        property_information['Living area'] = re.search(living_area_pattern2,extracted_text2).group(1)
    else:
        property_information['Living area'] = None
    
    #secondary_area_pattern
    secondary_area_pattern = r'kerrosala (\d+)m2'
    property_information['Secondary area'] = re.search(secondary_area_pattern,extracted_text2).group(1) if re.search(secondary_area_pattern,extracted_text2) else None
    
    #building year pattern
    building_year_pattern1 = r'Rakennusvuosi (\d{4})'
    building_year_pattern2 = r'(?<=\b)(\d{4}) (?=valmistunut)'

    if re.search(building_year_pattern1,extracted_text2):
        property_information['Building year'] = re.search(building_year_pattern1,extracted_text2).group(1)
    elif re.search(building_year_pattern2,extracted_text2):
        property_information['Building year'] = re.search(building_year_pattern2,extracted_text2).group(1)
    else:
        property_information['Building year'] = None
        
    #valuation_date_pattern
    for i in reversed(extracted_text):
        valuation_date_pattern = re.search(r'(\d{1,2}\.\d{1,2}\.\d{4})', i)
        if valuation_date_pattern:
            property_information['Valuation date'] = valuation_date_pattern.group(1)
    #         last_date = match.group(1)
            day, month, year = re.split(r'[.,-]', property_information['Valuation date'])
            property_information['Valuation date'] = f"{day}/{month}/{year}"
            break  # Stop searching once a date is found
        else:
            property_information['Valuation date'] = None
    
    
#     market_value_pattern = r'(\d+ ?\d*) €'
#     market_value_pattern = r'(\d+ ?\d*) euroa"
    market_value_pattern1 = r'(\d+(?: ?\d*)*) €'
    market_value_pattern2 = r'(\d+(?: ?\d*)*) euroa'
    for i in extracted_text2:
        if re.search(market_value_pattern1,extracted_text2):
            property_information['Market value'] = re.search(market_value_pattern1,extracted_text2).group(1).replace(' ', '')
        elif re.search(market_value_pattern2,extracted_text2):
            property_information['Market value'] = re.search(market_value_pattern2,extracted_text2).group(1).replace(' ', '')
        else:
            property_information['Market value'] = None
        
    #Rooms Pattern
    rooms_pattern1 = r'(\d+)h,'
    rooms_pattern2 = r'yksiö'
    rooms_pattern3 = r'kaksio'
    rooms_pattern4 = r'kolmio'
    rooms_pattern5 = r'Neljä'

    if re.search(rooms_pattern2,extracted_text2):
        property_information['Number of rooms'] = 1
    elif re.search(rooms_pattern3,extracted_text2):
        property_information['Number of rooms'] = 2
    elif re.search(rooms_pattern4,extracted_text2):
        property_information['Number of rooms'] = 3
    elif re.search(rooms_pattern5,extracted_text2):
        property_information['Number of rooms'] = 4
    elif re.search(rooms_pattern1,extracted_text2):
        property_information['Number of rooms'] = re.search(rooms_pattern1,extracted_text2).group(1)
    else:
        property_information['Number of rooms'] = None
    
    
    #primary_heating_source_pattern
    primary_heating_source_pattern = r'Lämmitysmuotona\s+on\s+([^,]+)'
    property_information['Primary heating source'] = re.search(primary_heating_source_pattern,extracted_text2).group(1) if re.search(primary_heating_source_pattern,extracted_text2) else None
    
    #condition_pattern
    condition_pattern1 = r'hyvässä'
    condition_pattern2 = r'hyväkunt'
    condition_pattern3 = r'Tyydyttävässä'
    condition_pattern4 = r'Tyydyttäväkunt'
    condition_pattern5=  r'Erinomainen'
    condition_pattern6=  r'Erinomaisessa'
    condition_pattern7=  r'Heikko'
    condition_pattern8=  r'Huono'


    if re.search(condition_pattern1,extracted_text2) or re.search(condition_pattern2,extracted_text2):
        property_information['Condition'] = 'Good condition'
    elif re.search(condition_pattern3,extracted_text2) or re.search(condition_pattern4,extracted_text2):
        property_information['Condition'] = 'Satisfactory'
    elif re.search(condition_pattern5,extracted_text2) or re.search(condition_pattern6,extracted_text2):
        property_information['Condition'] = 'Excellent'
    elif re.search(condition_pattern7,extracted_text2) or re.search(condition_pattern8,extracted_text2):
        property_information['Condition'] = 'Poor'
    else:
        property_information['Condition'] = None
    
    #Agent Name
    agent_name_pattern = r'Kiinteistönvälittäjä\s+(\w+)'
    property_information['External appraiser name'] = re.search(agent_name_pattern,extracted_text2).group(1) if re.search(agent_name_pattern, extracted_text2) else None
    

    return property_information
```
